Checking the type of data distribution for the columns which has continuous values using subplots as

```python
c=1
columns=[name for name in data.select_dtypes(include="float")]
for i in columns:
    plt.figure(figsize=(15,10))
    plt.subplot(3,3,c)
    sns.histplot(data[i])
    plt.xlabel(i)
    c+=1
```

```python
c=1
columns=[name for name in data.select_dtypes(include="float")]
for i in columns:
    plt.figure(figsize=(15,10))
    plt.subplot(3,3,c)
    data[i].plot.kde()
    plt.xlabel(i)
    c+=1
```

From this group of subplots we can observe that,

i) Maximum all the columns are following normal distribution.

ii) The columns which are following normal distribution with left skewness are collegeGPA, Domain, conscientiousness, agreeableness, extraversion, openess_to_experience.

iii) The  columns which are following normal distribution with no left and right skewness are 10percentage, 12percentage, nueroticism.

```python
columns=[name for name in data.select_dtypes(include=["int","float"])]
c=1
for i in columns:
    plt.figure(figsize=(15,10))
    plt.subplot(3,9,c)
    sns.boxplot(data[i],orient="v")
    c+=1
```

Finding the Outliers using IQR method as

```python
columns=[name for name in data.select_dtypes(include=["int","float"])]
lst=[]
for i in columns:
    Q1=data[i].quantile(0.25)
    Q3=data[i].quantile(0.75)
    IQR=Q3-Q1
    LL=Q1-(1.5*IQR)
    UL=Q3+(1.5*IQR)
    lst.append(data[(data[i]<LL)|(data[i]>UL)][i])
```

```python
for i in range(0,27):
    print(lst[i])
```

Above are the Outlier values in each column.
